relplot():

This is a figure-level-function that makes use of two other axes functions for Visualizing Statistical Relationships which are:

    scatterplot()
    lineplot()

These functions can be specified using the ‘kind’ parameter of relplot(). In case this parameter is given, it takes the default one which is scatterplot(). Before you begin writing your code, make sure to import the required libraries as follows:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
f = sns.load_dataset("flights") 
sns.relplot(x="passengers", y="month", data=f);
```

As you can see, the points are plotted in 2-dimensions. However, you can add another dimension using the ‘hue’ semantic. Let’s take a look at an example of the same:

```python
f = sns.load_dataset("flights") 
sns.relplot(x="passengers", y="month", hue="year", data=f);
```

```python
# Changing Color
sns.set(style="darkgrid")
f = sns.load_dataset("flights")
sns.relplot(x="passengers", y="month", hue="year",palette="ch:r=-.5,l=.75", data=f);
```

lineplot():

This function will allow you to draw a continuous line for your data. You can use this function by changing the ‘kind’ parameter as follows:

```python
a=pd.DataFrame({'Day':[1,2,3,4,5,6,7],'Grocery':[30,80,45,23,51,46,76],'Clothes':[13,40,34,23,54,67,98],'Utensils':[12,32,27,56,87,54,34]},index=[1,2,3,4,5,6,7]) 
g = sns.relplot(x="Day", y="Clothes", kind="line", data=a) 
g.fig.autofmt_xdate()
```

catplot():

This is a figure-level-function like relplot(). It can be characterized by three families of axes level functions namely:

    Scatterplots — These include stripplot(), swarmplot()
    Distribution Plots — which are boxplot(), violinplot(), boxenplot()
    Estimateplots — namely pointplot(), barplot(), countplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
a = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", data=a, kind='box');
```

Plotting bivariate distributions:

This comes into picture when you have two random independent variables resulting in some probable event. The best function to plot these type of graphs is jointplot(). Let us now plot a bivariate graph using jointplot().

```python
x=pd.DataFrame({'Day':[1,2,3,4,5,6,7],'Grocery':[30,80,45,23,51,46,76],'Clothes':[13,40,34,23,54,67,98],'Utensils':[12,32,27,56,87,54,34]},index=[1,2,3,4,5,6,7])
y=pd.DataFrame({'Day':[8,9,10,11,12,13,14],'Grocery':[30,80,45,23,51,46,76],'Clothes':[13,40,34,23,54,67,98],'Utensils':[12,32,27,56,87,54,34]},index=[8,9,10,11,12,13,14])
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="kde", color="b");
```

```python
# import seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt

# load the built-in tips dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# a regression line with a confidence inetrval plot
sns.lmplot(x='total_bill', y='tip', data = tips)
plt.show()
```

Now if we want to have two regression lines in one chart – one for female tips and one for male tips – we can modify the code above and use the hue parameter.

```python
sns.lmplot(x='total_bill',y='tip', data = tips, hue = 'sex')
plt.show()
```

On the other hand, we could also have one plot for each of the regression lines and write only one line of code.

```python
sns.lmplot(x='total_bill',y='tip', data = tips, col = 'sex')
plt.show()
```

```python
# Residual Plot
sns.residplot(x='total_bill',y='tip', data = tips)
plt.show()
```

```python
# STRIP PLOT
sns.stripplot(x='day',y ='tip',data = tips, size =5, jitter = True)

# overwrite the original y-label from 'tip' to 'tip ($)'
plt.ylabel('tip ($)')
plt.show()
```

```python
# SWARM PLOT
sns.swarmplot(x='day',y='tip',data = tips, hue = 'sex')

# overwrite the original y-label from 'tip' to 'tip ($)'
plt.ylabel('tip ($)')
plt.show()
```

```python
# BOX & VIOLIN PLOT
# create a grid with 1 row 2 columns
plt.subplot(1,2,1)

# the first plot
sns.boxplot(x='day',y='tip', data = tips)
plt.ylabel('tip ($)')

plt.subplot(1,2,2)

# the second plot
sns.violinplot(x='day',y='tip', data = tips)
plt.ylabel('tip ($)')
plt.tight_layout()
plt.show()
```

```python
# A joint plot with histograms
sns.jointplot(x='total_bill', y='tip', data = tips)
plt.show()
```

```python
# A joint plot with density estimates:
sns.jointplot(x='total_bill', y='tip', data = tips, kind = 'kde')
plt.show()
```

```python
# Pair Plot
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue = 'sex')
plt.show()
```

```python
# Heat Map
# compute correlations between features
df_corr = tips.corr()

# plot the correlations
sns.heatmap(df_corr)
plt.title('Correlation plot')
plt.show()
```
