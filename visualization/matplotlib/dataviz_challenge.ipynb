```python
import pandas as pd
# import plotly.graph_objects as go 
import matplotlib.pyplot as plt
%matplotlib inline


```

```python
df = pd.read_csv('results.csv')
df.head()
```

## Task
Which teams scored the largest number of goals in FIFA World Cup?

```python
df[df.tournament == 'FIFA World Cup' ].nlargest(5,['home_score','away_score'])
new_names = {'home_team':'team', 'away_team':'team', 'home_score':'score', 'away_score':'score'}
new_df = pd.concat( 
    [df[['home_team','home_score']].rename(columns = new_names)
    ,df[['away_team','away_score']].rename(columns = new_names)
    ], axis = 0
    )
ndf = new_df.groupby('team').sum().sort_values('score', ascending=False).head(10)

# https://www.shanelynn.ie/bar-plots-in-python-using-pandas-dataframes/
# KEY LESSON, BAR CHART REQUIRES VALUES AS SERIES WITH THE NAMES IN THE INDEX
# ndf.plot(kind='barh') # WORKS
fig, ax = plt.subplots(figsize = (10,6))
ax.bar(x = ndf.index,height = ndf.score,orientation='horizontal')
```

## Task
What is the number of matches played in each tournament throughout history?

```python
fig, ax = plt.subplots(figsize=(5, 10))
ax.barh(df.tournament.unique(), df.tournament.value_counts(), orientation='horizontal') #nbins
# ax.hist(df.tournament.unique(), orientation='horizontal')
# ax.hist()
plt.show()
```
