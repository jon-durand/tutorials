```python
# import pandas
import pandas as pd

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv')
```

```python
df.head()
```

# Objective: Read Total profit of all months and show it using a line plot

```python
plt.style.use('seaborn-colorblind')
x= df.month_number.values
y= df.total_profit.values
plt.plot(x,y, 'g',label='line one',linewidth=5)
plt.title('Total Profits By Month')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

# Objective: Read all product sales data and show it  using a multiline plot

```python
df.columns
```

```python
plt.style.use('seaborn-colorblind')
x= df.month_number.values
y1 = df.facecream.values
y2 = df.facewash.values
y3 = df.toothpaste.values
y4 = df.bathingsoap.values
y5 = df.shampoo.values
y6 = df.moisturizer.values

plt.plot(x,y1,label='facecream',linewidth=2)
plt.plot(x,y2,label='facewash',linewidth=2)
plt.plot(x,y3,label='toothpaste',linewidth=2)
plt.plot(x,y4,label='bathingsoap',linewidth=2)
plt.plot(x,y5,label='shampoo',linewidth=2)
plt.plot(x,y6,label='moisturizer',linewidth=2)

plt.title('Sales By Product')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend(loc='upper left')
plt.grid(True,color='k')
plt.show()
```

# Objective: Read the total profit of each month and show it using the histogram to see most common profit ranges

```python
# Histogram
x= df.month_number.values
y= df.total_profit.values
plt.hist(y, histtype='bar', rwidth=0.8)
plt.title('Histogram')
plt.xlabel('month')
plt.ylabel('total_profits')
plt.show()
```
